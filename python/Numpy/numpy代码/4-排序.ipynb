```python
import numpy as np
tang_array = np.array([[1.5,1.3,7.5],
                      [5.6,7.8,1.2]])
tang_array
```

```python
np.sort(tang_array) # 默认先从小到大排序
```

```python
np.sort(tang_array,axis = 0) # 指定轴排序——列
```

```python
np.sort(tang_array,axis = 1)
```

```python
tang_array
```

```python
np.argsort(tang_array) # 排序索引顺序
```

```python
tang_array = np.linspace(0,10,10) # 从0到10，元素个数
tang_array
```

```python
values = np.array([2.5,6.5,9.5])
```

```python
np.searchsorted(tang_array,values) # 排序插入，返回插入位置索引
```

```python
tang_array = np.array([[1,0,6],
                       [1,7,0],
                       [2,3,1],
                       [2,4,0]])
tang_array
```

```python
index = np.lexsort([-1*tang_array[:,0],tang_array[:,2]]) # 第0列降序，第2列升序
index
```

```python
tang_array = tang_array[index]
tang_array
```
