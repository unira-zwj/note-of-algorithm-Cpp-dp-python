```python
import numpy as np
import matplotlib
matplotlib.use('nbagg')
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
x = np.arange(5)
y = np.random.randint(-5,5,5)
print (y)
fig,axes = plt.subplots(ncols = 2)
v_bars = axes[0].bar(x,y,color='red')
h_bars = axes[1].barh(x,y,color='red')
plt.figure(4,4)

axes[0].axhline(0,color='grey',linewidth=2) # æ·»åŠ 0è½´çº¿
axes[1].axvline(0,color='grey',linewidth=2) # ç«–çº¿
plt.show()
```

```python
fig,ax = plt.subplots()
v_bars = ax.bar(x,y,color='lightblue')
for bar,height in zip(v_bars,y):
    if height < 0:
        bar.set(edgecolor = 'darkred',color = 'green')
ax.axhline(0, color = 'black', linewidth = 3)
plt.show()
```

```python
x = np.random.randn(100).cumsum() # cumsumæ±‚å’Œ
y = np.linspace(0,10,100)

fig,ax = plt.subplots()
ax.fill_between(x,y,color='lightblue')
plt.show()
```

```python
x = np.linspace(0,10,200)
y1 = 2*x +1
y2 = 3*x +1.2
y_mean = 0.5*x*np.cos(2*x) + 2.5*x +1.1
fig,ax = plt.subplots()
ax.fill_between(x,y1,y2,color='red')
ax.plot(x,y_mean,color='black')
plt.show()
```

```python
mean_values = [1,2,3]
variance = [0.2,0.4,0.5] # æŒ‡å®šè¯¯å·®èŒƒå›´
bar_label = ['bar1','bar2','bar3'] # æŸ±åç§°

x_pos = list(range(len(bar_label))) # xè½´ä½ç½®
plt.bar(x_pos,mean_values,yerr=variance,alpha=0.3) # ç»˜åˆ¶
max_y = max(zip(mean_values,variance)) 
plt.ylim([0,(max_y[0]+max_y[1])*1.2]) # yè½´é™åˆ¶
plt.ylabel('variable y') # yè½´åç§°
plt.xticks(x_pos,bar_label) # è®¾ç½®xè½´
plt.show()
```

```python
x1 = np.array([1,2,3])
x2 = np.array([2,2,3])

bar_labels = ['bat1','bar2','bar3']
fig = plt.figure(figsize = (6,4))
y_pos = np.arange(len(x2))
y_pos = [x for x in y_pos] # æŸ±çš„ä½ç½®

plt.barh(y_pos,x1,color='g',alpha = 0.5)
plt.barh(y_pos,-x2,color='b',alpha = 0.5)

plt.xlim(-max(x2)-1,max(x1)+1)
plt.ylim(-1,len(x1))
plt.show()
```

```python
green_data = [1, 2, 3]
blue_data = [3, 2, 1]
red_data = [2, 3, 3]
labels = ['group 1', 'group 2', 'group 3']

pos = list(range(len(red_data))) # æŸ±çš„ä½ç½®
width = 0.2 
fig, ax = plt.subplots(figsize=(6,4))

plt.bar(pos,green_data,width,alpha=0.5,color='g',label=labels[0])
plt.bar([p+width for p in pos],blue_data,width,alpha=0.5,color='b',label=labels[1])
plt.bar([p+width*2 for p in pos],red_data,width,alpha=0.5,color='r',label=labels[2])
plt.show()
```

```python
data = range(200, 225, 5)

bar_labels = ['a', 'b', 'c', 'd', 'e']

fig = plt.figure(figsize=(6,4))

y_pos = np.arange(len(data))

plt.yticks(y_pos, bar_labels, fontsize=12) # å®šä¹‰yè½´åˆ»åº¦

bars = plt.barh(y_pos, data, alpha = 0.5, color='g') # barhæ¨ªç€ç”»

plt.vlines(min(data), -1 ,len(data),linestyle = 'dashed') # ä»¥æœ€å°å€¼ç”»ç«–çº¿

# æ·»åŠ æ ‡æ³¨ï¼Œæ˜¯æœ€å°å€¼çš„ç™¾åˆ†ä¹‹å¤šå°‘å€
for b,d in zip(bars,data):
    plt.text(b.get_width()+b.get_width()*0.05,b.get_y()+b.get_height()/2,'{0:.2%}'.format(d/min(data)))
plt.show()
```

```python
mean_values = range(10,18)
x_pos = range(len(mean_values))

import matplotlib.colors as col
import matplotlib.cm as cm # color map

cmap1 = cm.ScalarMappable(col.Normalize(min(mean_values),max(mean_values),cm.hot))
cmap2 = cm.ScalarMappable(col.Normalize(0,20,cm.hot))

plt.subplot(121)
plt.bar(x_pos,mean_values,color = cmap1.to_rgba(mean_values))
plt.title('cmap1')

plt.subplot(122)
plt.bar(x_pos,mean_values,color = cmap2.to_rgba(mean_values))
plt.title('cmap2')

plt.show()
```

```python
patterns = ('-', '+', 'x', '\\', '*', 'o', 'O', '.')

fig = plt.gca()

mean_value = range(1,len(patterns)+1)
x_pos = list(range(len(mean_value)))

bars = plt.bar(x_pos,mean_value,color='green',alpha = 0.5)

for bar,pattern in zip(bars,patterns):
    bar.set_hatch(pattern)
plt.show()
len(patterns)
```
