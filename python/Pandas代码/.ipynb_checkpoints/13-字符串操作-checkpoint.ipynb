```python
import pandas as pd
import numpy as np
s = pd.Series(['A','b','B','gaer','AGER',np.nan]) # Series字符串属性
s
```

```python
s.str.lower() # 小写
```

```python
s.str.upper() # 大写
```

```python
s.str.len() # 查看长度
```
