```python
import pandas as pd
data = pd.DataFrame({'group':['a','a','a','b','b','b','c','c','c'],
                    'data':[4,3,2,1,12,3,4,5,7]})
data
```

```python
# by=['group','data'],ascending = [False,True] 横坐标以group降序的前提下再data升序
# inplace对原始数据操作
data.sort_values(by=['group','data'],ascending = [False,True],inplace=True)
data
```

```python
data = pd.DataFrame({'k1':['one']*3+['two']*4,
                   'k2':[3,2,1,3,3,4,4]})
data
```

```python
data.sort_values(by='k2') # 以k2值排序
```

```python
data.drop_duplicates() # 以k1k2同时重复的元素去重复
```

```python
data.drop_duplicates(subset='k1') # 以k1为索引去重复
```

```python
data = pd.DataFrame({'food':['A1','A2','B1','B2','B3','C1','C2'],'data':[1,2,3,4,5,6,7]})
data
```

```python

def food_map(series):
    if series['food'] == 'A1':
        return 'A'
    elif series['food'] == 'A2':
        return 'A'
    elif series['food'] == 'B1':
        return 'B'
    elif series['food'] == 'B2':
        return 'B'
    elif series['food'] == 'B3':
        return 'B'
    elif series['food'] == 'C1':
        return 'C'
    elif series['food'] == 'C2':
        return 'C'
# apply，对data数据的每个样本都执行food_map函数
data['food_map'] = data.apply(food_map,axis = 'columns') 
data
```

```python
food2Upper = {
    'A1':'A',
    'A2':'A',
    'B1':'B',
    'B2':'B',
    'B3':'B',
    'C1':'C',
    'C2':'C'
}
data['upper'] = data['food'].map(food2Upper)
data
```

```python
# 离散化
ages = [15,18,20,21,22,34,41,52,63,79]
bins = [10,40,80] # 10-40，40-80两个区间
bins_res = pd.cut(ages,bins) # ages带切分数据，bin3切分指定，如写5默认均分5份
bins_res
```

```python
bins_res.labels # 调属性
```

```python
pd.value_counts(bins_res) # 统计结果
```

```python
pd.cut(ages,[10,30,50,80])
```

```python
group_names = ['Yonth','Mille','Old']
#pd.cut(ages,[10,20,50,80],labels=group_names)
pd.value_counts(pd.cut(ages,[10,20,50,80],labels=group_names)) # labels给指定bin区间命名
```
