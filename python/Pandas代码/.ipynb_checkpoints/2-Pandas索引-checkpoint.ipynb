### Pandas索引结构

```python
import pandas as pd

df = pd.read_csv('./data/titanic.csv')
```

```python
df['Age'][:5]
```

```python
df[['Age','Fare']][:5]
```

* loc 用label来去定位，可通过内容取
* iloc 用position来去定位，通过索引取

```python
df.iloc[0]
```

```python
df.iloc[0:5]
```

```python
df.iloc[0:5,1:3]
```

```python
df = df.set_index('Name')
```

```python
df.loc['Heikkinen, Miss. Laina']
```

```python
df.loc['Heikkinen, Miss. Laina','Fare']
```

```python
df.loc['Heikkinen, Miss. Laina':'Allen, Mr. William Henry',:]
```

```python
df.loc['Heikkinen, Miss. Laina','Fare'] = 1000
df.head()
```

```python
#要被遗弃了，别用了
df.ix['Heikkinen, Miss. Laina','Fare']
```

### bool类型的索引

```python
df['Fare'] > 40
```

```python
df[df['Fare'] > 40][:5]
```

```python
df[df['Sex'] == 'male'][:5]
```

```python
df.loc[df['Sex'] == 'male','Age'].mean() # 男性乘客年龄取平均
```

```python
(df['Age'] > 70).sum()
```
